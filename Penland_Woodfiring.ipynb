```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
woodfiring_df = pd.read_csv('/Users/kaitokikuchi/Dropbox/Personal/Penland/Penland_Woodfiring_Temp_2024.csv')
woodfiring_df['Time'] = pd.to_datetime(woodfiring_df['Time'])


plt.plot(woodfiring_df['Time'], woodfiring_df['Front Temperature'], label='Front', c='orange', lw=2)
plt.plot(woodfiring_df['Time'], woodfiring_df['Middle Temperature'], label='Middle', c='red', lw=2)
plt.plot(woodfiring_df['Time'], woodfiring_df['Back Temperature'], label='Back', c='purple', lw=2)
plt.plot(woodfiring_df['Time'], woodfiring_df['Target Temperature'], label='Target', c='k', ls=':', lw=2)




plt.xlabel('Time', fontsize=15)
plt.ylabel('Temperature (F)', fontsize=15)
plt.xticks(rotation=45)
plt.legend();
```
